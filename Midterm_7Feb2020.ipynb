## 1 > สร้าง function ชื่อ smartDict ซึ่งทำหน้าที่
- function รับ input เป็น list 2 ตัว ที่มีขนาดเท่ากัน โดย list แรกเป็น list ของตัวเลข list ที่สองเป็น list ของ text
- ถ้าค่าใน list แรก เป็นเลขคี่ ให้ค่าใน dict ที่เป็น output มี key เป็นค่าจาก list ตัวที่ 1 และ value จาก list ตัวที่ 2
- ถ้าค่าใน list แรก เป็นเลขคู่ ให้ค่าใน dict ที่เป็น output มี key เป็นค่าจาก list ตัวที่ 2 และ value จาก list ตัวที่ 1
- output ของ function คือ dict ที่สร้างขึ้น


```python
def smartDict(listnum, listtext):
    if listnum:
        retdict = {}
        for index in range(len(listnum)):         
            retdict[listnum[index]] = listtext[index]
    else:
        retdict = {}
        for index in range(len(listnum)):         
            retdict[listtext[index]] = listnum[index]
    return retdict
```

```python
smartDict ([1,2,3],['apple', 'banana', 'orange'])
```

## 2 > จากไฟล์ Wuhan_nCoV_GNome.txt จงเขียนโปรแกรมเพื่อตอบคำถามต่อไปนี้

### 2.1 > กรดอะมิโน (A, T, C, G) มีปริมาณมากที่สุดและมีเท่าไร

```python
import pandas as pd
```

```python
DNAseq = pd.read_csv('C:\\Users\\MUBU\\Documents\\Basic Data Science\\Wuhan_nCoV_GNome.txt', header = None)
DNAseq.head()
```

```python
def countChar(seq, character):
    charcount = 0
    for char in seq:
        if char == character:
            charcount = charcount + 1
    return charcount
```

```python
ANum = 0
TNum = 0
GNum = 0
CNum = 0
for row in range(DNAseq.shape[0]):
    ANum += countChar (DNAseq.iloc[row][0], 'A')
    TNum += countChar (DNAseq.iloc[row][0], 'T')
    GNum += countChar (DNAseq.iloc[row][0], 'G')
    CNum += countChar (DNAseq.iloc[row][0], 'C')
print(f'There are {ANum} As, {TNum} Ts, {GNum} Gs, {CNum} Cs')
```

### 2.2 > จงหา sequence ทที่มีขนาด 3 และ 4 ที่เกิดขึ้นซ้ำมากที่สุด

```python
def populate3charstring(seq):
    str_list = list()
    for i in range(len(seq)-2):
        str_list.append(seq[i:i+3])
    return str_list
```

```python
list_all_3char = list()
for row in range(DNAseq.shape[0]):
    list_all_3char = list_all_3char + populate3charstring (DNAseq.iloc[row][0])
len(list_all_3char)
```

```python
unique_3char = list(set(list_all_3char))
len(unique_3char)
```

```python
dict_3char = {}
for sub_str in unique_3char:
    num = 0
    for s in list_all_3char:
        if sub_str == s:
            num += 1
    dict_3char[sub_str] = num
```

```python
dict_3char
```

```python
maxVal = -100
maxKey = None
for k in dict_3char.keys():
    if dict_3char[k] > maxVal:
        maxVal = dict_3char[k]
        maxKey = k
maxKey
```

```python
def populate4charstring(seq):
    str_list = list()
    for i in range(len(seq)-3):
        str_list.append(seq[i:i+4])
    return str_list
```

### 3.2 > สร้างตาราง Dictionary โดยตารางประกอบด้วย 2 columns คือ ภาษาอีสาน และภาษาไทย โดยค่าใน column ภาษาอีสาน คือ คำในภาษาอีสาน และ column ภาษาไทย คือ คำแปลเป็นภาษากลางของภาษาอีสานใน column แรก

```python
labels = ['ภาษาอีสาน' , 'ภาษาไทย']
df2 = pd.DataFrame.from_records([],columns = labels)
for i in range(32):
    data_f2 = pd.read_excel('C:\\Users\\MUBU\\Documents\\Basic Data Science\\data.xlsx', sheet_name = i, usecols = 'B:C', skipfooter = 1)
    df2 = pd.concat([df2,data_f2], axis = 0, ignore_index = True)
```

```python
df2
```

```python
dict_match = ['ภาษาอีสาน = ภาษาไทย']
for i in range(df2.shape[0]) :
    ES_spt = df2.iloc[i]['ภาษาอีสาน'].split('/')
    TH_spt = df2.iloc[i]['ภาษาไทย'].split('/')
    if len(ES_spt) == len(TH_spt):
        match_spt = len(ES_spt)
        for u in range(match_spt):
            dict_match.append(f'{ES_spt[u]} = {TH_spt[u]}')
```
