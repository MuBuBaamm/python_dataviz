```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
corona = pd.read_csv('C:\\Users\\MUBU\\Downloads\\2019_nCoV_data.csv')
corona.head()
```

```python
corona.shape
```

```python
corona.isnull().any()
```

```python
dataset = corona.dropna(subset = ['Province/State']).sort_values('Confirmed', ascending=False)
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isnull().any()
```

# Bar chart

```python
coRC = dataset['Recovered'].groupby([dataset['Country']]).sum()
coRC.head()
```

```python
table_coRC = pd.DataFrame(coRC)
table_coRC.head()
```

```python
coCR.plot(kind='bar',x='Country',y='Recovered')
plt.show()
```

# Line chart

```python
coCD = dataset['Confirmed'].groupby([dataset['Date']]).sum()
coCD.head()
```

```python
coDD = dataset['Deaths'].groupby([dataset['Date']]).sum()
coDD.head()
```

```python
table_coCD = pd.DataFrame(coCD)
table_coCD.head()
```

```python
table_coDD = pd.DataFrame(coDD)
table_coDD.head()
```

```python
coCD.plot(kind='line',x='Confirmed',y='Date')
plt.show()
```

```python
coDD.plot(kind='line',x='Deaths',y='Date')
plt.show()
```

```python
ax = plt.gca() 
dataset.plot(kind='line',x='Date',y='Confirmed',ax=ax) 
dataset.plot(kind='line',x='Date',y='Deaths', color='red', ax=ax) 
plt.show() 
```

# Pie chart

# Area chart

```python
plt.stackplot(dataset.index,
              [dataset['Confirmed'], dataset['Deaths'], dataset['Recovered']],
              labels=['Confirmed', 'Deaths', 'Recovered'],
              alpha=0.8)

plt.legend(loc=2, fontsize='large')
plt.show()
```
