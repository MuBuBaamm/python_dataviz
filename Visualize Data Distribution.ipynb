```python
irisdata = pd.read_csv('C:\\Users\\MUBU\\Documents\\Basic Data Science\\iris.data', header = None)
irisdata.head()
```

```python
setosa_np = irisdata.iloc[:50, :-1].to_numpy()
```

```python
versicolor_np = irisdata.iloc[50:100, :-1].to_numpy()
```

```python
virginica_np = irisdata.iloc[100:, :-1].to_numpy()
```

```python
matplotlib.rcParams['figure.figsize'] = [12,12]
```

```python
plt.subplot(4, 2, 1)
plt.scatter(setosa_np[:,1], setosa_np[:,2], s=30*setosa_np[:,3], c=setosa_np[:,0], cmap='PuRd', marker='8') 
plt.scatter(versicolor_np[:,1], versicolor_np[:,2], s=30*versicolor_np[:,3], c=versicolor_np[:,0], cmap='PuRd', marker='d')
plt.scatter(virginica_np[:,1], virginica_np[:,2], s=30*virginica_np[:,3], c=virginica_np[:,0], cmap='PuRd', marker='_')

plt.subplot(4, 2, 2)
plt.scatter(setosa_np[:,2], setosa_np[:,3], s=30*setosa_np[:,0], c=setosa_np[:,1], cmap='PuRd', marker='8') 
plt.scatter(versicolor_np[:,2], versicolor_np[:,3], s=30*versicolor_np[:,0], c=versicolor_np[:,1], cmap='PuRd', marker='d')
plt.scatter(virginica_np[:,2], virginica_np[:,3], s=30*virginica_np[:,0], c=virginica_np[:,1], cmap='PuRd', marker='_') 

plt.subplot(4, 2, 3) 
plt.scatter(setosa_np[:,3], setosa_np[:,0], s=30*setosa_np[:,1], c=setosa_np[:,2], cmap='PuRd', marker='8') 
plt.scatter(versicolor_np[:,3], versicolor_np[:,0], s=30*versicolor_np[:,1], c=versicolor_np[:,2], cmap='PuRd', marker='d')
plt.scatter(virginica_np[:,3], virginica_np[:,0], s=30*virginica_np[:,1], c=virginica_np[:,2], cmap='PuRd', marker='_') 

plt.subplot(4, 2, 4) 
plt.scatter(setosa_np[:,0], setosa_np[:,1], s=30*setosa_np[:,2], c=setosa_np[:,3], cmap='PuRd', marker='8') 
plt.scatter(versicolor_np[:,0], versicolor_np[:,1], s=30*versicolor_np[:,2], c=versicolor_np[:,3], cmap='PuRd', marker='d')
plt.scatter(virginica_np[:,0], virginica_np[:,1], s=30*virginica_np[:,2], c=virginica_np[:,3], cmap='PuRd', marker='_')

plt.subplot(4, 2, 5)
plt.scatter(setosa_np[:,1],setosa_np[:,0],s=30*setosa_np[:,2], c = setosa_np[:,3], cmap='PuRd', marker='8')
plt.scatter(versicolor_np[:,1],versicolor_np[:,0],s=30*versicolor_np[:,2], c = versicolor_np[:,3], cmap='PuRd', marker='d')
plt.scatter(virginica_np[:,1],virginica_np[:,0],s=30*virginica_np[:,2], c = virginica_np[:,3], cmap='PuRd', marker='_')

plt.subplot(4, 2, 6)
plt.scatter(setosa_np[:,2],setosa_np[:,1],s=30*setosa_np[:,0], c = setosa_np[:,3], cmap='PuRd', marker='8')
plt.scatter(versicolor_np[:,2],versicolor_np[:,1],s=30*versicolor_np[:,0], c = versicolor_np[:,3], cmap='PuRd', marker='d')
plt.scatter(virginica_np[:,2],virginica_np[:,1],s=30*virginica_np[:,0], c = virginica_np[:,3], cmap='PuRd', marker='_')

plt.subplot(4, 2, 7)
plt.scatter(setosa_np[:,3],setosa_np[:,1],s=30*setosa_np[:,0], c = setosa_np[:,2], cmap='PuRd', marker='8')
plt.scatter(versicolor_np[:,3],versicolor_np[:,1],s=30*versicolor_np[:,0], c = versicolor_np[:,2], cmap='PuRd', marker='d')
plt.scatter(virginica_np[:,3],virginica_np[:,1],s=30*virginica_np[:,0], c = virginica_np[:,2], cmap='PuRd', marker='_')

plt.subplot(4, 2, 8)
plt.scatter(setosa_np[:,1],setosa_np[:,3],s=30*setosa_np[:,2], c = setosa_np[:,0], cmap='PuRd', marker='8')
plt.scatter(versicolor_np[:,1],versicolor_np[:,3],s=30*versicolor_np[:,2], c = versicolor_np[:,0], cmap='PuRd', marker='d')
plt.scatter(virginica_np[:,1],virginica_np[:,3],s=30*virginica_np[:,2], c = virginica_np[:,0], cmap='PuRd', marker='_')
```
