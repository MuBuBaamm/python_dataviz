# Boxplot

```python
import pandas as pd
```

```python
irisdata = pd.read_csv('C:\\Users\\MUBU\\Documents\\Basic Data Science\\iris.data', header = None)
```

```python
setosa_np = irisdata.iloc[:50, :-1].to_numpy()
versicolor_np = irisdata.iloc[50:100, :-1].to_numpy()
virginica_np = irisdata.iloc[100:, :-1].to_numpy()
```

```python
### อยากทราบว่า plot boxplot ของ setosa_np ได้กี่ boxplot
setosa_np
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [12,12]              # plot ให้เรียงกัน 4 อัน แนวนอน ซึ่ง setosa มีด้วยกัน 4 column จึง plot ได้ 4 boxplot
plt.subplot(1, 4, 1)
plt.boxplot(setosa_np[:,0])
plt.subplot(1, 4, 2)
plt.boxplot(setosa_np[:,1])
plt.subplot(1, 4, 3)
plt.boxplot(setosa_np[:,2])
plt.subplot(1, 4, 4)
plt.boxplot(setosa_np[:,3])
plt.show()
```

```python
plt.boxplot(setosa_np)                       # plot รวมกัน
```

```python
Debt_df_debt = Debt_df[['Region and province', 2549]]         # เก็บ column ที่สนในไว้ในไฟล์ใหม่
```

```python
Debt_df_debt.drop([0, 1, 2, 3, 8, 31, 49, 70, 85, 86, 87, 88], axis = 0, inplace = True)         # ลบแถวที่ไม่เกี่ยวข้องออก
```

```python
Debt_df_debt.head()
```

```python
data_to_plot = Debt_df_debt         # สร้างไฟล์ใหม่ให้เก็บตารางที่ลบแถวที่ไม่เกี่ยวข้องออกแล้ว
```

```python
data_to_plot.head()
```

```python
import json
```

```python
with open('C:\\Users\\MUBU\\Documents\\Basic Data Science\\thai.json') as json_file:
    country_geo = json.load(json_file)
```

```python
new_col = list()                                                          # ลบช่องว่างข้างหน้าชื่อออกเพื่อที่จะสามารถทำให้ไฟล์ 2 ไฟล์นั้นสามารถเชื่อมกันได้
for d in data_to_plot['Region and province']:
    new_col.append(d.lstrip())
data_to_plot['Region and province'] = new_col                 
```

```python
map = folium.Map(location = [10,100], zoom_start=d)          # กำหนดพิกัด ลองติจูดและละติจูดของไทย
```

```python
map
```

```python
# ลงสีของข้อมูลที่มีค่าของจำนวนหนี้สินเฉลี่ยต่อครัวเรือน
# ถ้าจังหวัดไหนมีจำนวนหนี้สินเฉลี่ยต่อครัวเรือนมากจะเป็นสีม่วงเข้ม
# ส่วนจังหวัดไหนมีจำนวนหนี้สินเฉลี่ยต่อครัวเรือนน้อยจะเป็นสีเหลือง ตามแถบรูปข้างบน
map.choropleth(geo_data = country_geo,
               data = data_to_plot,
               columns = ['Region and province', 2549],
               key_on = 'feature.properties.name',
               fill_color = 'YlGnBu', fill_opacity = 0.7,
               line_opacity = 0.2,
               legend_name = 'Average Debt')
```

```python
map.save('plot_data.html')
```
